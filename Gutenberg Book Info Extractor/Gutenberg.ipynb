```python
ebooks = soup.find('table', class_='files').find_all('tr')

formats = []
links = []
sizes = []

for row in ebooks:
    for a, col in enumerate(row.find_all('td')):
#         print(f"{a}: , {col.text}")
        if a==1:
            formats.append(col.text)
        elif a == 2:
            links.append(col.text)
        elif a == 3:
            sizes.append(col.text)   
```

## Final Function :

```python
def get_information(url):
    site = requests.get(url)
    soup = BeautifulSoup(site.text, 'html.parser')
    
    ebooks = soup.find('table', class_='files').find_all('tr')

    formats = []
    links = []
    sizes = []

    for row in ebooks:
        for a, col in enumerate(row.find_all('td')):
    #         print(f"{a}: , {col.text}")
            if a==1:
                formats.append(col.text)
            elif a == 2:
                links.append(col.text)
            elif a == 3:
                sizes.append(col.text)   
                
    Downloads = pd.DataFrame({"Format":formats, "Links":links, "Sizes":sizes})
    
    
    bio = soup.find('table', class_='bibrec').find_all('tr')

    ths =[]
    tds=[]
    for row in bio:
        try :
    #         print(f"{row.th.text} : {row.td.text}")
            ths.append(row.th.text)
            tds.append(row.td.text)

        except:
            continue
    Bibliographic = pd.DataFrame(tds, index=ths)
    
    return(Downloads, Bibliographic)
        
    
```
